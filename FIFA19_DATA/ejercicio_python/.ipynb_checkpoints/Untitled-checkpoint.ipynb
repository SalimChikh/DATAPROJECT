```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('../data/data.csv')
```

```python
dataset.drop('Loaned From', inplace=True, axis=1)
dataset.drop('Contract Valid Until', inplace=True, axis=1)
dataset.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
release_clause = list(dataset['Release Clause'].values)
weights = list(dataset['Weight'].values)
```

```python
release_int = list()
for money in release_clause:
    money_number = float(money[1:-1])*10**6
    release_int.append(money_number)
```

```python
weight_int = list()
for w in weights:
    weight_number = int(w[:-3])
    weight_int.append(weight_number)
```

```python
dataset['Weight'] = weight_int
dataset['Release Clause'] = release_int
```

```python
dataset_num = dataset.select_dtypes(np.number)
```

```python
dataset_num.set_index('ID', inplace=True)
```

```python
dataset_num.head(10)
```

```python
dataset_num.columns
```

```python
bad_variables = ['Weak Foot', 'Skill Moves', 'Jersey Number', 'Release Clause']
dataset_num.drop(bad_variables, axis=1, inplace=True)
```

```python
dataset_num.columns
```

```python
dataset_num.shape
```

```python
dataset_num.to_csv('../data/numeric_data.csv', sep=',')
```
